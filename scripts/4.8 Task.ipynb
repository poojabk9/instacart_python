### Q4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

### Q6. Create a spending flag for each user based on the average price across all their orders using the following criteria:

#### 1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender".

```python
# Calculate the mean prices for each user

df_ords_prods_merge['user_mean_prices'] = df_ords_prods_merge.groupby('user_id')['prices'].mean()
```

```python
df_ords_prods_merge.head()
```

```python
# Create a spending flag based on the criteria "Low spender"

df_ords_prods_merge.loc[df_ords_prods_merge['user_mean_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.head()
```

#### 2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create a spending flag based on the criteria "High spender"

df_ords_prods_merge.loc[df_ords_prods_merge['user_mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Q7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Calculating median user orders and creating 'user_median_order' column

df_ords_prods_merge['user_median_order'] = df_ords_prods_merge.groupby('user_id')['days_since_prior_order'].median()
```

```python
df_ords_prods_merge.head()
```



#### 1. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”


```python
df_ords_prods_merge.loc[df_ords_prods_merge['user_median_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.head()
```

#### 2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['user_median_order'] > 10) & (df_ords_prods_merge['user_median_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.head()
```

#### 3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
df_ords_prods_merge.loc[df_ords_prods_merge['user_median_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
