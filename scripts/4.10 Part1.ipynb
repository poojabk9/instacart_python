### Q3. The Instacart officers are interested in comparing customer behavior in different geographic areas. Create a regional segmentation of the data. You’ll need to create a “Region” column based on the “State” column from your customers data set.

```python
# Checking all the states

df_cust_ords['state'].unique()
```

```python
# Creating a region variable for each state

result = []

for value in df_cust_ords["state"]:
    if value in ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']:
        result.append('NorthEast')
    elif value in ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']:
        result.append('MidWest')
    elif value in ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']:
        result.append('South')
    else:
        result.append('West')
```

```python
result
```

```python
# Creating a new column 'region' and assigning the values in 'result' variable created above

df_cust_ords['region'] = result
```

```python
df_cust_ords['region'].value_counts(dropna = False)
```

```python
# Grouping by region and checking the mean price

df_cust_ords.groupby('region').agg({'prices': ['mean']})
```
