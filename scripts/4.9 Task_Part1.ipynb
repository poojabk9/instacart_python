### Q4. Wrangle the data so that it follows consistent logic; for example, rename columns with illogical names and drop columns that don’t add anything to your analysis.

```python
df_cust.columns
```

```python
# Renaming columns

df_cust.rename(columns = {'First Name' : 'first_name'}, inplace = True)
df_cust.rename(columns = {'Surnam' : 'last_name'}, inplace = True)
df_cust.rename(columns = {'Gender' : 'gender'}, inplace = True)
df_cust.rename(columns = {'STATE' : 'state'}, inplace = True)
df_cust.rename(columns = {'Age' : 'age'}, inplace = True)
df_cust.rename(columns = {'n_dependants' : 'dependants'}, inplace = True)
df_cust.rename(columns = {'fam_status' : 'family_status'}, inplace = True)
```

```python
df_cust.columns
```

```python
# Updating Datatypes

df_cust.dtypes
```

```python
df_cust['user_id'] = df_cust['user_id'].astype('str')
df_cust['date_joined'] = pd.to_datetime(df_cust['date_joined'])
```

### Q5. Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# Cheking null values

df_cust.isnull().sum()
```

```python
# Removing null values

df_nan = df_cust[df_cust['first_name'].isnull() == True]
```

```python
df_nan
```

```python
df_cust_clean = df_cust[df_cust['first_name'].isnull() == False]
```

```python
df_cust_clean.shape
```

```python
# Checking duplicates

df_dups = df_cust_clean[df_cust_clean.duplicated()]
```
