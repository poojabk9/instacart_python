### Q7. Aggregate the max, mean, and min variables on a customer-profile level for usage frequency and expenditure.

```python
df_merged.loc[df_merged['prices'] >100, 'prices'] = np.nan
```

```python
aggregated_data = df_merged.groupby('profile_1').agg({
    'order_number': ['max', 'mean', 'min'],
    'prices': ['max', 'mean', 'min']
    })
```

### Q8. Compare your customer profiles with regions and departments. Can you identify a link between them? Customers may demonstrate different behaviors across regions, which could make this information important.

```python
contingency_table = pd.crosstab(df_merged['profile_1'], [df_merged['region'], df_merged['department']])
```

```python
# Plot a heatmap
sns.heatmap(contingency_table, annot=True, cmap='viridis', fmt='d')
```

```python
# Create a bar chart

plt.figure(figsize=(10, 6))
df_merged.groupby(['region', 'profile_1']).size().unstack().plot(kind='bar', stacked=True)

# Set plot labels and title

plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customer Profiles Across Regions')
```

```python
plt.savefig(os.path.join(path, 'Analysis','Visualization', 'Customer_Profiles_Across_region.png'))
```

```python
# Create a bar chart

plt.figure(figsize=(10, 6))
df_merged.groupby(['department', 'profile_1']).size().unstack().plot(kind='bar', stacked=True)

# Set plot labels and title

plt.xlabel('Department')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customer Profiles Across department')
```

```python
plt.savefig(os.path.join(path, 'Analysis','Visualization', 'Customer_Profiles_Across_department.png'))
```
